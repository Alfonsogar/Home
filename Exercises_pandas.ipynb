# Comandos avanzados

### Convertir mediante una función la duración en horas

```python
def conv_time (duration):
    hours= int(duration/60)
    minutes = int (duration%60)
    
    return "%i h %i minutes" % (hours, minutes)
```

```python
movies['duration_h_m'] = movies.duration.apply(lambda x:conv_time(x))
```
